# SVR

# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

# Read World bank CSV

```python
csv_path = 'WorldBank.csv'
```

```python
data = pd.read_csv(csv_path)

print(data.head(10))
```

# Perform conditional selection to find - Population ages 15-64 (% of total population)

```python
output = data[data['Indicator Name'] == 'Population ages 15-64 (% of total population)']
print("Output for ages between 15-64:: ")
print(output)
```

# Divide the data into input and output

```python
x = np.arange(1960,2020).reshape(-1,1)
y = output.values[0][4:-1]

print(x)
print(y)
```

# Plot scatter plot of Population ages 15-64 (% of total population)

```python
plt.scatter(x, y)
plt.xlabel('Years')
plt.ylabel('Percentage')
```

# Splitting the dataset into the Training set and Test set

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.7, random_state=7)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
print(x_train)
```

```python
print(y_train)
```

# Fitting SVR on 3 Different Kernel on dataset

```python
model_linear = SVR(kernel='linear')
```

```python
model_poly = SVR(kernel='poly', C=1.0, epsilon=0.1)
```

```python
model_rbf = SVR(kernel='rbf')
```

```python
model_sig = SVR(kernel='sigmoid')
```

```python
model_linear.fit(x_train, y_train)
```

```python
model_poly.fit(x_train, y_train)
```

```python
model_rbf.fit(x_train, y_train)
```

```python
model_sig.fit(x_train, y_train)
```

# Predict the x_test using 3 Kernel

```python
y_predict_linear = model_linear.predict(x_test)
y_predict_poly = model_poly.predict(x_test)
y_predict_rbf = model_rbf.predict(x_test)
y_predict_sig = model_sig.predict(x_test)
```

# Visualising the  results

```python
plt.scatter(x_test, y_test)
plt.scatter(x_test, y_predict_linear)
plt.scatter(x_test, y_predict_rbf)
plt.scatter(x_test, y_predict_sig)
```

# Decision Tree Classifier

# Import iris data ser using sklearn

```python
from sklearn.datasets import load_iris
```

# Importing the libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

# Divide the data into input and output

```python
data = load_iris()
```

```python
x = data.data
```

```python
y = data.target
```

# Splitting the dataset into the Training set and Test set

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(x,y,test_size=0.3,random_state=4)
print(x1_test)
```

# Fitting DecisionTreeClassifier   on dataset

```python
model = DecisionTreeClassifier()
```

```python
t = model.fit(x1_train, y1_train)
```

# Display Decision Tree

```python
tree.plot_tree(t)
```

# Predict the x_test 

```python
y_predict = model.predict(x1_test)
```

```python
accuracy_score(y1_test, y_predict)
```
