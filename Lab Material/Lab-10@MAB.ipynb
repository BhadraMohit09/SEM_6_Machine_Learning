# Step 1: Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import AgglomerativeClustering, DBSCAN, KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
```

# Step 2: Generate dummy data using make_blobs

```python
data, labels = make_blobs(n_samples=1000, centers=850, cluster_std=1.0, random_state=42)
```

# Step 3: Plot Generated data 

```python
plt.figure(figsize=(10, 7))

plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='gist_rainbow')

plt.title('Scatter plot of the data')

plt.show()
```

# Step 4: Apply Hierarchical Clustering

```python
hierarchical = AgglomerativeClustering(n_clusters=3)
```

```python
hierarchical.fit(data)
```

```python
plt.figure(figsize=(10, 7))

plt.scatter(data[:, 0], data[:, 1], c=hierarchical.labels_, cmap='rainbow')

plt.title('Scatter plot of the data')
```
